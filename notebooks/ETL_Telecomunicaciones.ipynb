```python
# Importación de librerías
import pandas as pd
import numpy as np

# Rutas de los archivos
path_internet = 'C:/Users/Usuario/Desktop/DATA SCIENCE/PROYECTO_DA/data/raw/Internet.xlsx'
path_mapa_conectividad = 'C:/Users/Usuario/Desktop/DATA SCIENCE/PROYECTO_DA/data/raw/mapa_conectividad.xlsx'
path_telefonia_fija = 'C:/Users/Usuario/Desktop/DATA SCIENCE/PROYECTO_DA/data/raw/telefonia_fija.xlsx'
path_telefonia_movil = 'C:/Users/Usuario/Desktop/DATA SCIENCE/PROYECTO_DA/data/raw/Telefonia_movil.xlsx'

# Carga de hojas específicas del dataset Internet
internet_tecnologia_localidad = pd.read_excel(path_internet, sheet_name='Accesos_tecnologia_localidad')
internet_totales_tecnologia = pd.read_excel(path_internet, sheet_name='Totales Accesos Por Tecnología')
internet_accesos_tecnologia = pd.read_excel(path_internet, sheet_name='Accesos Por Tecnología')

# Carga de otros datasets
mapa_conectividad = pd.read_excel(path_mapa_conectividad)
telefonia_fija = pd.read_excel(path_telefonia_fija)
telefonia_movil = pd.read_excel(path_telefonia_movil)

# Exploración inicial de cada DataFrame
print("Internet Tecnología por Localidad")
print(internet_tecnologia_localidad.head())
print(internet_tecnologia_localidad.info())

print("\nTotales Accesos Por Tecnología")
print(internet_totales_tecnologia.head())
print(internet_totales_tecnologia.info())

print("\nAccesos Por Tecnología")
print(internet_accesos_tecnologia.head())
print(internet_accesos_tecnologia.info())

print("\nMapa de Conectividad")
print(mapa_conectividad.head())
print(mapa_conectividad.info())

print("\nTelefonía Fija")
print(telefonia_fija.head())
print(telefonia_fija.info())

print("\nTelefonía Móvil")
print(telefonia_movil.head())
print(telefonia_movil.info())
```

```python
# Limpieza de 'internet_tecnologia_localidad'
internet_tecnologia_localidad['Accesos'] = internet_tecnologia_localidad['Accesos'].fillna(0).astype(int)

internet_tecnologia_localidad.head()
```

```python
# Limpieza de 'internet_totales_tecnologia'
internet_totales_tecnologia['Año'] = internet_totales_tecnologia['Año'].astype('category')
internet_totales_tecnologia['Trimestre'] = internet_totales_tecnologia['Trimestre'].astype('category')

internet_totales_tecnologia.head()
```

```python
# Limpieza de 'internet_accesos_tecnologia'
internet_accesos_tecnologia = internet_accesos_tecnologia.fillna(0)
internet_accesos_tecnologia[['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros', 'Total']] = \
    internet_accesos_tecnologia[['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros', 'Total']].astype(int)

internet_accesos_tecnologia.head()
```

```python

# Limpieza de 'mapa_conectividad'
mapa_conectividad.replace({'SI': True, '--': False}, inplace=True)
mapa_conectividad[['ADSL', 'Cablemódem', 'Dial Up', 'Fibra óptica', 'Satelital', 'Wireless', 'Telefonía Fija', '3G', '4G']] = \
    mapa_conectividad[['ADSL', 'Cablemódem', 'Dial Up', 'Fibra óptica', 'Satelital', 'Wireless', 'Telefonía Fija', '3G', '4G']].astype(bool)
mapa_conectividad['Link'] = mapa_conectividad['Link'].fillna(0)

mapa_conectividad.head()
```

```python

# Limpieza de 'telefonia_fija'
telefonia_fija['Año'] = telefonia_fija['Año'].astype('category')
telefonia_fija['Trimestre'] = telefonia_fija['Trimestre'].astype('category')

telefonia_fija.head()
```

```python
# Limpieza de 'telefonia_movil'
telefonia_movil['Año'] = telefonia_movil['Año'].astype('category')
telefonia_movil['Trimestre'] = telefonia_movil['Trimestre'].astype('category')

telefonia_movil.head()
```

```python
# Limpieza de la columna 'Año' en 'internet_accesos_tecnologia'
internet_accesos_tecnologia['Año'] = internet_accesos_tecnologia['Año'].str.extract('(\d+)', expand=False).astype(float)

# Verificar si hay valores nulos después de la conversión
print("Valores nulos en 'Año' después de la limpieza:", internet_accesos_tecnologia['Año'].isnull().sum())
```

```python
# Rellenar nulos con un año específico (por ejemplo, el más común)
common_year = internet_accesos_tecnologia['Año'].mode()[0]
internet_accesos_tecnologia['Año'].fillna(common_year, inplace=True)
```

```python
# Asegurarse de que la columna 'Año' sea de tipo string para aplicar .str.extract
internet_accesos_tecnologia['Año'] = internet_accesos_tecnologia['Año'].astype(str)
internet_accesos_tecnologia['Año'] = internet_accesos_tecnologia['Año'].str.extract('(\d+)', expand=False).astype(float)

# Asegurarse de que la columna 'Trimestre' sea de tipo string para aplicar .str.extract
internet_accesos_tecnologia['Trimestre'] = internet_accesos_tecnologia['Trimestre'].astype(str)
internet_accesos_tecnologia['Trimestre'] = internet_accesos_tecnologia['Trimestre'].str.extract('(\d+)', expand=False).astype(float)

# Eliminar filas donde 'Año' o 'Trimestre' sean nulos
internet_accesos_tecnologia = internet_accesos_tecnologia.dropna(subset=['Año', 'Trimestre'])

# Convertir las columnas 'Año' y 'Trimestre' a tipo entero, si ya no contienen valores nulos
internet_accesos_tecnologia['Año'] = internet_accesos_tecnologia['Año'].astype(int)
internet_accesos_tecnologia['Trimestre'] = internet_accesos_tecnologia['Trimestre'].astype(int)

```

```python
import os
import pandas as pd

# Obtener la ruta absoluta de la carpeta actual
base_path = os.getcwd()
processed_path = os.path.join(base_path, 'data', 'processed')

# Verificar si la carpeta "data/processed" existe; si no, crearla
if not os.path.exists(processed_path):
    os.makedirs(processed_path)

# Limpieza y conversión de columnas problemáticas en cada DataFrame

# 1. Limpieza de 'internet_tecnologia_localidad'
internet_tecnologia_localidad['Link Indec'] = pd.to_numeric(internet_tecnologia_localidad['Link Indec'], errors='coerce')

# Convertir las columnas de tipo 'object' a tipo string si es necesario
internet_tecnologia_localidad['Provincia'] = internet_tecnologia_localidad['Provincia'].astype(str)

# 2. Limpieza de 'internet_accesos_tecnologia'
# Convertir 'Provincia' a string en 'internet_accesos_tecnologia'
internet_accesos_tecnologia['Provincia'] = internet_accesos_tecnologia['Provincia'].astype(str)

# Limpieza de la columna 'Año' y 'Trimestre' en 'internet_accesos_tecnologia'
internet_accesos_tecnologia['Año'] = internet_accesos_tecnologia['Año'].astype(str).str.extract('(\d+)', expand=False).astype(float)
internet_accesos_tecnologia['Trimestre'] = internet_accesos_tecnologia['Trimestre'].astype(str).str.extract('(\d+)', expand=False).astype(float)

# Eliminar filas con valores nulos en 'Año' o 'Trimestre' tras la conversión
internet_accesos_tecnologia = internet_accesos_tecnologia.dropna(subset=['Año', 'Trimestre'])

# Convertir 'Año' y 'Trimestre' a tipo entero
internet_accesos_tecnologia['Año'] = internet_accesos_tecnologia['Año'].astype(int)
internet_accesos_tecnologia['Trimestre'] = internet_accesos_tecnologia['Trimestre'].astype(int)

# Guardado de los DataFrames en formato Parquet
internet_tecnologia_localidad.to_parquet(os.path.join(processed_path, 'internet_tecnologia_localidad.parquet'))
internet_totales_tecnologia.to_parquet(os.path.join(processed_path, 'internet_totales_tecnologia.parquet'))
internet_accesos_tecnologia.to_parquet(os.path.join(processed_path, 'internet_accesos_tecnologia.parquet'))
mapa_conectividad.to_parquet(os.path.join(processed_path, 'mapa_conectividad.parquet'))
telefonia_fija.to_parquet(os.path.join(processed_path, 'telefonia_fija.parquet'))
telefonia_movil.to_parquet(os.path.join(processed_path, 'telefonia_movil.parquet'))

```
